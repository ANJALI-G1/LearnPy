```python
wishlist_anjali = {"Japan", "Italy", "Brazil", "Norway", "India", "Japan"}
wishlist_friend = {"India", "Canada", "Brazil", "Germany", "Italy"}


# 1Ô∏è‚É£ Remove duplicates from both wishlists
wishlist_anjali=set(wishlist_anjali)
wishlist_friend=set(wishlist_friend)

print(wishlist_anjali)
print(wishlist_friend)


# 2Ô∏è‚É£ Find countries both want to visit

wishlist_common=wishlist_anjali.intersection(wishlist_friend)
print(wishlist_common)
# 3Ô∏è‚É£ Find countries only Anjali wants to visit
wishlist_anjali_only=wishlist_anjali-wishlist_common
print(wishlist_anjali_only)
# 4Ô∏è‚É£ Find countries either one wants to visit, but not both
wishlist_single=wishlist_anjali.union(wishlist_friend)-wishlist_common
print(wishlist_single)
# 5Ô∏è‚É£ Add "New Zealand" to Anjali's wishlist
wishlist_anjali.add("New Zealand")
print(wishlist_anjali)
# 6Ô∏è‚É£ Remove "Germany" from friend's wishlist safely
wishlist_friend.remove("Germany")
# 7Ô∏è‚É£ Check if "Norway" is in friend's wishlist
if "Norway" in wishlist_friend:
    print("present")
else:
    print("Not present")
```

```python
# üß≥ Bonus: Travel Budget Filter

# Each country has an estimated cost in USD
country_costs = {
    "Japan": 1200,
    "Italy": 1000,
    "Brazil": 800,
    "Norway": 1500,
    "India": 500,
    "Canada": 1100,
    "Germany": 950,
    "New Zealand": 1300
}

# 1Ô∏è‚É£ Filter Anjali's wishlist to countries costing less than $1000
l=[]
for i in wishlist_anjali:
    if(country_costs[i]<1000):
        l.append(i)
print(l)
# 2Ô∏è‚É£ Create a sorted list of all countries (from both wishlists) by cost
wishlist_anjali=list(wishlist_anjali)
print(wishlist_anjali)
wishlist_anjali=sorted(wishlist_anjali)
print(wishlist_anjali)

wishlist_friend=list(wishlist_friend)
print(wishlist_friend)

wishlist_friend=sorted(wishlist_friend)
print(wishlist_friend)

anjali_cost=[  country_costs[i] for i in wishlist_anjali  ]
print(anjali_cost)

friend_cost=[  country_costs[i] for i in wishlist_friend ]
print(friend_cost)

# 3Ô∏è‚É£ Find the most expensive country in the combined wishlist
max_country=max(wishlist_common,key=lambda x: country_costs[x])
print(max_country)
l=[]
for i in wishlist_common:
   l.append(country_costs[i])
print(l)
print(max(l))
```

```python
sorted_shared = sorted(wishlist_common, key=lambda x: country_costs[x])
for country in sorted_shared:
    print(f"{country}: ${country_costs[country]}")
```

```python
#‚úÖ 1. Create a dictionary
student = {
    "name": "Anjali",
    "age": 21,
    "courses": ["Python", "Math"],
    "marks": {"Python": 95, "Math": 88}
}
# ‚úÖ 2. Access values using keys
# Print the student's name and age.
print(student["name"],student["age"])

# Print the marks in Python.
for i in student["marks"]:
    print()

# Write a function add_student() that takes a student ID and details, and adds it to the students dictionary.
def addId(id,detail):
    student.update({id:detail})
addId(112,"roll")
print(student)

# ‚úÖ 3. Add new key-value pairs
# Add "email": "anjali@example.com"
student.update({"email":"anjali@example.com"})
print(student)

# Add a new course "English" with marks 90
student["marks"].update({"English":90})
print(student["marks"])

# ‚úÖ 4. Update existing values
# Update age to 22'
student["age"]=22
print(student)

# Update marks in Math to 92
print(student["marks"])
student["marks"]["Math"]=92
print(student["marks"]["Math"])

# ‚úÖ 5. Delete a key
# Remove the "email" key using del
del(student["email"])
print(student)

# ‚úÖ 6. Use dictionary methods
# keys(), values(), items()

# Loop through marks and print subject-wise scores.
for i  in student["marks"]:
    print( i,student["marks"][i])

key_=student.keys()
print(key_)
values_=student.values()
print(values_)

for i in student:
    print(i,student[i])

# ‚úÖ 7. Check if a key exists
if "English" in student["marks"]:
    print("English marks:", student["marks"]["English"])

# 8. Use get() method
print(student.get("phone", "Not Available"))
print(student.get("age", "Not Available"))


# ‚úÖ 9. Nested dictionary
# Create a dictionary of multiple students:
students = {
    "101": {"name": "Anjali", "age": 22},
    "102": {"name": "Rahul", "age": 23}
}

print(students["102"]["age"])

```

```python
# Create a student record system
students = {}

def add_student(student_id, name, age, courses, marks):
    students[student_id] = {
        "name": name,
        "age": age,
        "courses": courses,
        "marks": marks
    }

def get_average(student_id):
    if student_id in students:
        marks = students[student_id]["marks"]
        return sum(marks.values()) / len(marks)
    return "Student not found"

# Add a student
add_student("101", "Anjali", 22, ["Python", "Math", "English"], {"Python": 95, "Math": 92, "English": 90})
print("Average Marks:", get_average("101"))
```
